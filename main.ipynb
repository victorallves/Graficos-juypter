```python
import pandas as pd
import matplotlib as plt
```

```python

confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
grafico = confirmed.groupby("Country/Region").sum()
grafico.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
grafico.loc["China"][3:10].plot()
plt.title("Casos confirmados China")
```

```python
plt.figure(figsize=(10,8))
grafico.loc["Brazil"][100:110].plot(kind='bar')
plt.title("Casos confirmado no Brasil")
plt.show()
```

```python
plt.figure(figsize=(12,5))
data = "3/2/20"
grafico[data].sort_values(ascending=False)[1:10].plot(kind='pie')
plt.title("Os paises com mais casos confirmados na data (3/2/2020) tirando a China (primeiro lugar)")
plt.show()
```
